```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.flipkart.com/search?q=gaming%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
response=requests.get(url)
print("The response is positive :",response.status_code)#request is sucessfull it shows 200
```

```python
soup=BeautifulSoup(response.text,"html.parser")
```

```python
Product_Name=[]
Ratting=[]
offers=[]
sizes=[]
colors=[]
weights=[]
discount=[]
prices=[]
for pages in range(1,502):
    print(f"scrapping....{pages}")
    url=f"https://www.flipkart.com/search?q=gaming+laptop&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_2_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_2_2_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=gaming+laptop&requestId=eaf49150-573a-4a7c-b48e-fab8533dbaa8&page={pages}"
    response=requests.get(url)
    soup=BeautifulSoup(response.text,"html.parser")
    for g in soup.find_all("span",class_="Wphh3N"):
        Ratting.append(g.text.strip() if g else 0)
    for m in soup.find_all("a", class_="wjcEIp"):
        Product_Name.append(m.text.strip() if m else None)
    for offer in soup.find_all("div",class_="yiggsN O5Fpg8"):
        offers.append(offer.text.strip() if offer else None)
    for price in soup.find_all("div",class_="Nx9bqj"):
        prices.append(price.text.strip() if price else None)
    for dis in soup.find_all("span",string= lambda x: x and x.endswith('% off')):
        discount.append(dis.text.strip() if dis else None)
    for specs in soup.find_all("div", class_="NqpwHC"):
        text = specs.get_text(strip=True)
        parts = [part.strip() for part in text.split(',')]
        
        size = parts[0] if len(parts) > 0 else None 
        color = parts[1] if len(parts) > 1 else None
        weight = next((part for part in parts if "kg" in part.lower()), 0)

        sizes.append(size)
        colors.append(color)
        weights.append(weight)
```

```python
print("product_names:",len(Product_Name))
print(" offer :",len(offers))
print("Ratting available:",len(Ratting))
print("Discount:",len(discount))
print("prices :",len(prices))
print("sizes :",len(sizes))
print("colors :",len(colors))
print("weights :",len(weights))
```

```python
min_lenght=min(len(Product_Name),len(weights),len(colors),len(sizes),len(Ratting),len(prices),len(discount))
import pandas as  pd
data={
    "Laptop Name":Product_Name[:min_lenght],
    "weights":weights[:min_lenght],
    "Colors":colors[:min_lenght],
    "sizes":sizes[:min_lenght],
    "Rating":Ratting[:min_lenght],
    "Discount":discount[:min_lenght],
    "Prices":prices[:min_lenght]}
gamming_data=pd.DataFrame(data)
import numpy as np
gamming_data.replace({None: np.nan}, inplace=True)
gamming_data.to_csv("gamming_data.csv", index=False)

print("Data saved as 'gamming_data.csv'")
```
